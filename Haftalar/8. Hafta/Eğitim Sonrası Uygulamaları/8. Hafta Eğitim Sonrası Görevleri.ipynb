# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
# 1.Problemin Çözümünü Buraya Yazınız
def vkiÖlçer():
    print("Vücüt Kitle İndeksi Ölçer'e Hoş Geldin!")
    print()
    turSayisi=int(input("Kaç kere giriş yapacaksınız, kaç ölçüm yapacaksınız? "))
    for i in range(1,turSayisi+1):
        class user():
            def __init__(self,name,surname,age,height,weight):
                self.name=name
                self.surname=surname
                self.age=age
                self.height=height
                self.weight=weight
                self.bmi=round(weight/height**2,1)
        info=user(input("İsminizi giriniz: "),input("Soyisminizi giriniz: "),input("Yaşınızı giriniz: "),float(input("Boyunuzu (metre olarak(Örn. 1.75)) giriniz: ")),float(input("Kilonuzu (kg) giriniz: ")))

        bmiClass=""
        bmiClasses=["Zayıf","Normal Kilolu","Fazla Kilolu","Obez","İleri Derecede Obez (Morbid Obez)"]
        if info.bmi<18.5:
            bmiClass=bmiClasses[0]
        elif 18.5<=info.bmi<=24.9:
            bmiClass=bmiClasses[1]
        elif 25<=info.bmi<=29.9:
            bmiClass=bmiClasses[2]
        elif 30<=info.bmi<=39.9:
            bmiClass=bmiClasses[3]
        elif 40<=info.bmi:
            bmiClass=bmiClasses[4]

        print(f"Vücüt Kitle İndeksiniz: {info.bmi} Vücüt Kitle İndeksi Sınıfınız: {bmiClass}")

        userContent=f"({info.name}, {info.surname}, {info.age}, {info.height}, {info.weight}, {bmiClass})\n"
        with open("VeriSeti.txt","a+") as file:
            file.write(userContent)
vkiÖlçer()
```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
# 2.Problemin Çözümünü Buraya Yazınız
import time

def YetBar():
    print("YetBar'a Hoş Geldin!")
    print()
    makineAçık=True
    class Icecek():
        def __init__(self,isim,tarih,icecekismi=""):
                self.isim=isim
                self.tarih=tarih
                self.icecekismi=""
                self.icerik=[]
                self.fiyat=0

    while makineAçık:
        option=input("Ne yapmak istersin? (Sayıyı giriniz) \n 1. Menüyü Görüntüleme \n 2. Smoothie Satın Alma \n 3. Makineyi Kapatma")
        while not(option.isdigit()):
            option=input("Lütfen doğru bir sayı giriniz. \n 1. Menüyü Görüntüleme \n 2. Smoothie Satın Alma \n 3. Makineyi Kapatma")
        
        menü={
            "Muz":12,
            "Çilek":14,
            "Ananas":15,
            "Elma":10,
            "Havuç":11,
            "Portakal":12
        }
        menüKeys=list(menü.keys())
        menüValues=list(menü.values())

        if option=="1":
            print("Menü: ")
            for i in range(0,6):
                print(f"{i+1}.{menüKeys[i]}->{menüValues[i]}")
            option=input("Başka bir şey yapmak ister misin? (Sayıyı giriniz) \n 1. Menüyü Görüntüleme \n 2. Smoothie Satın Alma \n 3. Makineyi Kapatma")
        
        if option=="2":
            isim=input("İsminiz nedir?")
            tarih=input("Lütfen bugünün tarihini (GG.AA.YYYY formatında) giriniz: ")
            icecek=Icecek(isim,tarih)
            def icerikSeç():
                while True:
                    seçim=input("Eklemek istediğiniz ürünlerin sırasıyla numarasını giriniz, bittiğinde çıkmak için 'q' tuşuna basınız.")
                    if seçim=="q":
                            break
                    elif seçim.isdigit():
                        seçim=int(seçim)
                        icecek.icerik.append(menüKeys[seçim-1])
                        icecek.fiyat+=menüValues[seçim-1]
                        if seçim=="q":
                            break
            icerikSeç()
            print()
            print("...İçeceğiniz Hazırlanıyor...")
            print()
            time.sleep(2.5)
            for i in range(0,len(icecek.icerik)):
                 icecek.icecekismi+=icecek.icerik[i][0:2]
            print(f"{icecek.icecekismi} Hazır! Ücretiniz: {icecek.fiyat}")
            data=f"({icecek.isim}, {icecek.tarih}, {icecek.icecekismi}, {icecek.icerik}, {icecek.fiyat})"
            with open("YetBarData.txt","a+") as file:
                file.write(f"{data}\n")
            option=input("Başka bir şey yapmak ister misin? (Sayıyı giriniz) \n 1. Menüyü Görüntüleme \n 2. Smoothie Satın Alma \n 3. Makineyi Kapatma")
        
        if option=="3":
            print()
            print("YetBar'ı kullandığınız için teşekkürler, iyi günler!")
            makineAçık=False
YetBar()
```
