# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
# 1.Problemin Çözümünü Buraya Yazınız
class vehicle():
    def __init__(self,model,colour,max_speed):
        self.model=model
        self.colour=colour
        self.max_speed=max_speed

class bus(vehicle):
    def __init__(self,model,colour,max_speed,seating_capacity=50):
        super().__init__(model,colour,max_speed)
        self.seating_capacity=seating_capacity

Bus=bus("Mercedes 210","Blue",160)
print(f"The seating capacity of a bus is {Bus.seating_capacity} passengers.")
````

````python
# 2.Problemin Çözümünü Buraya Yazınız
class school_bus():
    def __init__(self,model,colour,max_speed,seating_capacity):
        self.model=model
        self.colour=colour
        self.max_speed=max_speed
        self.seating_capacity=seating_capacity
School_bus=school_bus("Volkswagen","Yellow",110,25)

#school_bus burada kendi başına bir class'tır ve vehicle'dan da bus'dan da bağımsızdır. 
# Bu yüzden de School_bus vehicle classının da bus classının da bir instance'ı (örneği) değildir.

print(f"vehicleÖrnek-->{isinstance(School_bus,vehicle)}\nvehicleClass-->{issubclass(school_bus,vehicle)}\nbusÖrnek-->{isinstance(School_bus,bus)}\nbusClass-->{issubclass(school_bus,bus)}\n(Koddaki notları lütfen kontrol ediniz.)\n")

class school_bus(vehicle):
    def __init__(self,model,colour,max_speed,seating_capacity):
        super().__init__(model,colour,max_speed)
        self.seating_capacity=seating_capacity
School_bus=school_bus("Volkswagen","Yellow",110,25)

#school_bus burada sadece vehicle classının bir subclass'ıdır, bus classının değildir.
#vehicle classında olmayan seating_capacity attribute'na sahip olmasına rağmen vehicle'daki attribute'ların hepsini karşıladığı için vehicle classının örneği sayılır ve zaten bus classının subclass'ı olmadığı için onun direkt olarak örneği olmaz.

print(f"vehicleÖrnek-->{isinstance(School_bus,vehicle)}\nvehicleClass-->{issubclass(school_bus,vehicle)}\nbusÖrnek-->{isinstance(School_bus,bus)}\nbusClass-->{issubclass(school_bus,bus)}\n(Koddaki notları lütfen kontrol ediniz.)\n")

class school_bus(bus):
    def __init__(self,model,colour,max_speed,seating_capacity):
        super().__init__(model,colour,max_speed,seating_capacity)
School_bus=school_bus("Volkswagen","Yellow",110,25)

#school_bus burada bus classının subclassı durumundadır ve bus classı da zaten vehicle classının subclassı olduğu için school_bus da direkt olarak vehicle classının da subclassı olmuştur.
#Bu yüzden hem bus classının hem de daha fazla attribute'a sahip olduğu için yine vehicle classının örneği sayılır.

print(f"vehicleÖrnek-->{isinstance(School_bus,vehicle)}\nvehicleClass-->{issubclass(school_bus,vehicle)}\nbusÖrnek-->{isinstance(School_bus,bus)}\nbusClass-->{issubclass(school_bus,bus)}\n(Koddaki notları lütfen kontrol ediniz.)\n")
````
